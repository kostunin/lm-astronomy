```python
import pandas as pd
import json
import numpy as np
import re
```

```python
with open('entities.json') as f:
    data = json.load(f)

df = pd.DataFrame.from_dict(data, orient='index')

atel_no = df.index

coo0 = np.array(df['coordinates'])
#for i in range(len(coo0)):
#    print (coo0[i], '            atel_no', atel_no[i])
```

```python
file1 = open('coo_list.txt', 'w')

coo = [i for i, x in enumerate(coo0) if x is None]
atel_no1 = [value for i, value in enumerate(atel_no) if i not in coo]
coo = [x for x in coo0 if x is not None]

for i in range(len(coo)):
    if (len(coo[i]) == 1):
        print(coo[i], file = file1)
        print(coo[i], '    atel_no = ', atel_no1[i])
    else:
        length = len(coo[i])
        for j in range(length):
            print (coo[i][j], file = file1)
            print (coo[i][j], '    atel_no = ', atel_no1[i])
file1.close()
```

```python
def replace_word_in_file(file_path, old_word, new_word):
    # Read the contents of the file
    with open(file_path, 'r') as file:
        content = file.read()

    # Replace the old word with the new word
    content = content.replace(old_word, new_word)

    # Write the modified content back to the file
    with open(file_path, 'w') as file:
        file.write(content)
        
def enclose_rows_in_quotes(file_path):
    # Read the contents of the file
    with open(file_path, 'r') as file:
        lines = file.readlines()

    # Enclose each line between single quotes
    lines = ["'" + line.strip() + "'\n" for line in lines]

    # Write the modified content back to the file
    with open(file_path, 'w') as file:
        file.writelines(lines)

def find_number_with_sign(input_string):
    pattern = re.compile(r'[-+]?\d+[-+]\d+')

    match = pattern.search(input_string)

    if match:
        return match.group()
    else:
        return None

file_name = 'coo_list.txt'
replace_word_in_file(file_name, '[', '')
replace_word_in_file(file_name, ']', '')
replace_word_in_file(file_name, "'", '')
enclose_rows_in_quotes(file_name)

with open(file_name, 'r') as file:
    lines = file.readlines()
data = [' '.join(line.strip().split()) for line in lines]

counts = []
for i, string in enumerate(data, start=1):
    # Split the string by space and count the number of elements
    count = len(string.split())
    counts.append(count)

print ('Possible broken entries (lower number:)')
for i in range(len(data)):
    if (counts[i] != 2 and counts[i] != 6 and counts[i] != 4):
        result = find_number_with_sign(data[i])
        if result == None:
            print (data[i], '    atel_no = ',atel_no1[i])

```
